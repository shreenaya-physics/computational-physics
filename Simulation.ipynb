```python
import pandas as pd
import matplotlib.pyplot as plt
import glob


files = sorted(glob.glob("*.csv"))

for file in files:


    df = pd.read_csv(file, skiprows=3)
    df.columns = df.columns.str.replace('# ', '', regex=False)

    time = df["Time (s)"]

    plt.figure(figsize=(10,6))

    plt.plot(time, df["Altitude (m)"], label="Altitude")
    plt.plot(time, df["Vertical velocity (m/s)"], label="Vertical velocity")
    plt.plot(time, df["Total velocity (m/s)"], label="Total velocity")
    plt.plot(time, df["Lateral distance (m)"], label="Lateral distance")
    plt.plot(time, df["Angle of attack (°)"], label="Angle of attack")

    plt.xlabel("Time (s)")
    plt.ylabel("Value")
    plt.title(f"Simulation Results - {file}")
    plt.legend()
    plt.grid(True)

    plt.show()
```
